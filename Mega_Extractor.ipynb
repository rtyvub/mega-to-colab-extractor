```python
# Install a system-level MEGA downloader to bypass Python errors
!apt-get -qq install megatools

import os
import zipfile
import glob

# --- YOUR MEGA LINK ---
mega_url = 'YOUR_MEGA_LINK_HERE'
# ----------------------

extract_folder = '/content/Extracted_Photos'
os.makedirs(extract_folder, exist_ok=True)

print("Downloading ZIP to temporary self-destructing storage...")
# Download the file using the megatools system command
!megadl '{mega_url}' --path /content/

print("Unzipping files...")
# Find the downloaded zip file
zip_files = glob.glob('/content/*.zip')

if zip_files:
    downloaded_file = zip_files[0]
    with zipfile.ZipFile(downloaded_file, 'r') as zip_ref:
        zip_ref.extractall(extract_folder)

    # Delete the ZIP to save temp space
    os.remove(downloaded_file)
    print("✅ Done! Click the Folder icon on the far left sidebar, then open 'Extracted_Photos' to view your photos.")
else:
    print("❌ Error: Could not find the downloaded ZIP file.")
```
